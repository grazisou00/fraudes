```python
# Célula 1: Configuração e Carregamento
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, f1_score, recall_score, precision_score

# 1. Carregamento dos Dados
# O nome do arquivo deve ser exatamente 'creditcard.csv'
try:
    df = pd.read_csv('creditcard.csv')
    print("Dados carregados com sucesso.")
except FileNotFoundError:
    print("Erro: Arquivo 'creditcard.csv' não encontrado. Certifique-se de que o arquivo foi carregado no Colab.")
```

```python
# 2. Análise Exploratória e Pré-processamento

# 2.1. Verificação de Desbalanceamento (Crucial para o Artigo)
print("\nContagem de Classes (0: Não Fraude, 1: Fraude):")
print(df['Class'].value_counts())

# 2.2. Visualização do Desbalanceamento (Gerar Imagem para o Artigo)
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=df)
plt.title('Distribuição de Classes')
plt.savefig('distribuicao_classes.png') # Salve a imagem para o artigo
plt.show()

# 2.3. Separação de Features (X) e Target (y)
X = df.drop('Class', axis=1)
y = df['Class']

# 2.4. Divisão em Treino e Teste (70% Treino, 30% Teste)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"\nTamanho do conjunto de Treino: {X_train.shape[0]} amostras")
print(f"Tamanho do conjunto de Teste: {X_test.shape[0]} amostras")
```

```python
# 3. Modelagem e Treinamento

# 3.1. Inicialização do Modelo Random Forest
# class_weight='balanced' é a técnica de mitigação de desbalanceamento escolhida.
model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced', n_jobs=-1)

# 3.2. Treinamento do Modelo
print("\nIniciando treinamento do modelo Random Forest...")
model.fit(X_train, y_train)
print("Treinamento concluído.")
```

```python
# 4. Avaliação do Desempenho

# 4.1. Previsão no Conjunto de Teste
y_pred = model.predict(X_test)

# 4.2. Geração da Matriz de Confusão (Tabela 2)
cm = confusion_matrix(y_test, y_pred)
print("\nMatriz de Confusão:")
print(cm)

# 4.3. Geração do Relatório de Classificação (Tabela 1)
report = classification_report(y_test, y_pred, target_names=['Não Fraude', 'Fraude'])
print("\nRelatório de Classificação:")
print(report)

# 4.4. Extração das Métricas Chave para o Artigo
# Lembre-se de usar as métricas da CLASSE 1 (Fraude)
f1 = f1_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
accuracy = model.score(X_test, y_test)

print("\n--- Métricas Finais para o Artigo ---")
print(f"Acurácia Geral: {accuracy:.4f}")
print(f"Precisão (Fraude): {precision:.4f}")
print(f"Recall (Fraude): {recall:.4f}")
print(f"F1-Score (Fraude): {f1:.4f}")

# 4.5. Visualização da Matriz de Confusão (Gerar Imagem para o Artigo)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Não Fraude', 'Fraude'], yticklabels=['Não Fraude', 'Fraude'])
plt.title('Matriz de Confusão')
plt.ylabel('Real')
plt.xlabel('Previsto')
plt.savefig('matriz_confusao.png')
plt.show()
```
